### Dynamic Axes

```python
from bokeh.models import (
    Dropdown,
    ColumnDataSource,
)
from bokeh.plotting import figure
from bokeh.layouts import column
from hover.utils.bokeh_helper import servable

@servable()
def burner():
    df = pd.DataFrame({
        'f0': np.random.uniform(-1.0, 1.0, size=(100,)).tolist(),
        'f1': np.random.uniform(-1.0, 1.0, size=(100,)).tolist(),
        'f2': np.random.uniform(-1.0, 1.0, size=(100,)).tolist(),
        'f3': np.random.uniform(-1.0, 1.0, size=(100,)).tolist(),
    })
    choose_x = Dropdown(label="Choose X", menu=['f0', 'f1', 'f2', 'f3'])
    choose_y = Dropdown(label="Choose Y", menu=['f0', 'f1', 'f2', 'f3'])
    
    source = ColumnDataSource(df)
    plot = figure()
    renderer = plot.scatter(x='f0', y='f1', source=source)

    def change_x(event):
        renderer.glyph.x = event.item
        
    def change_y(event):
        renderer.glyph.y = event.item
    
    choose_x.on_click(change_x)
    choose_y.on_click(change_y)
    return column(choose_x, choose_y, plot)
```

```python
handle = burner()
show(handle)
```
